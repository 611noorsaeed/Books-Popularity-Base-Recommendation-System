```python
merged_with_name = books.merge(ratings,on='ISBN')
```

```python
merged_with_name = merged_with_name[['ISBN','Book-Title','Book-Author','Book-Rating','Image-URL-M']]
```

```python
merged_with_name
```

```python
num_ratings = merged_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings.rename(columns={"Book-Rating":'num-ratings'},inplace=True)
num_ratings
```

```python
avg_rating = merged_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'avg-ratings'},inplace=True)
avg_rating
```

```python
popular_df = num_ratings.merge(avg_rating,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num-ratings'] >=250].sort_values('avg-ratings',ascending=False).head(50)
popular_df
```
